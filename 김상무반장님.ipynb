```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_excel('부산경남경마장 안드로이드 비콘현황v1.3(부경본부)_수정.xlsx')
ModifyingData = pd.read_excel('부산경남경마공원(20210703).xlsx')
```

```python
howlong = int(len(ModifyingData["MAC Address"]))
k=0
while k<howlong-1:
  MACADDRESS = ModifyingData["MAC Address"][k]
  copy=-1;
  answer=-1
  number=0
  for i in data["5GHz"]:
    if i==MACADDRESS:
      answer=number#이름이 몇번째 열인지 answer에 저장
      copy=i#MACADDRESS 저장
    number=number+1  
  
  number=0 
  for e in data["2.4GHz"]:
    if e==MACADDRESS:
      answer=number
      copy=e
    number=number+1

   
  number=0
  for a in data["기기 MAC"]:
    if a==MACADDRESS:
      answer=number
      copy=a
    number=number+1

  if answer!=-1 :
    ModifyingData["내용"][k]=data["AP현황"][answer]
  k=k+1
```
